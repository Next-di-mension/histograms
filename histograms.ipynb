```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Define the dictionaries
counts = {"000000000": 53, "100000011": 467, "110000000": 46, "100000001": 4, "000000011": 45, "100000000": 440, "000000010": 70, "000000001": 63}
counts2 = {"000000000": 90, "100000011": 300, "110000000": 56, "100000001": 3, "000000011": 55, "100000000": 40, "000000010": 80, "000000001": 68}
counts3 = {"000000000": 190, "100000011": 30, "110000000": 576, "100000001": 13, "000000011": 155, "100000000": 4, "000000010": 280, "000000001": 8}
# Create a DataFrame from the dictionaries
df = pd.DataFrame({'Bitstring': list(counts.keys()), 'Counts 1': list(counts.values()), 'Counts 2': list(counts2.values()), 'Counts 3': list(counts3.values())})

# Melt the DataFrame to "long" format
df_melted = df.melt(id_vars='Bitstring', var_name='Dataset', value_name='Count')

# Set the size of the plot
sns.set(rc={'figure.figsize':(10,6)})

# Plot the barplot
sns.barplot(data=df_melted, x='Bitstring', y='Count', hue='Dataset', palette='Blues_d', edgecolor='.2')

# Add axis labels and a title
plt.xlabel('Bitstring')
plt.ylabel('Count')
plt.title('Histogram of Counts for Two Datasets')

# Add a legend
plt.legend(title='Dataset')

# Display the plot
plt.show()
```
